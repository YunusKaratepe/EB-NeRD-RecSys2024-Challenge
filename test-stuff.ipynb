```python
# Check gender column for non-NaN values
print("=== GENDER COLUMN ANALYSIS ===")

# Check if gender column exists in any of the dataframes
dataframes = {'articles': articles, 'history': history, 'behaviors': behaviors}

for name, df in dataframes.items():
    print(f"\n{name.upper()} DataFrame:")
    print(f"Columns: {df.columns.tolist()}")
    
    if 'gender' in df.columns:
        print(f"Gender column found!")
        print(f"Total rows: {len(df)}")
        print(f"Non-NaN gender values: {df['gender'].notna().sum()}")
        print(f"NaN gender values: {df['gender'].isna().sum()}")
        print(f"Unique gender values: {df['gender'].dropna().unique()}")
        print(f"Gender value counts:")
        print(df['gender'].value_counts(dropna=False))
    else:
        print("No 'gender' column found")
```
